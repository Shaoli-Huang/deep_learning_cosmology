```python
import theano
import numpy
import theano.tensor as T
from theano import function
rng = numpy.random

x = T.dscalar('x')
y = T.dscalar('y')
z = x + y 
f = function([x, y],z)
#f(2,3)

m = T.matrix('m')
n = T.matrix('n')
b = T.dscalar('b')
mult = m * n + b
fmat = function([m, n, b],mult)
#fmat([[1, 2, 3 ,4 ],[3, 5, 7, 8]], [[1, 2, 3 ,4 ],[3, 5, 7, 8]],1.0)
#fmat() 

a = T.vector()
out = a + a**2
fpow = function([a],out)
#fpow(numpy.array([1, 2, 3, 4, 5]))

#logistic
x_num = T.matrix('x_num')
s = 1/(1 + T.exp(-x_num))
flogr = function([x_num], s)
#flogr([[1, -1.293537], [-1.637429302, 0.02362784]])

#random var
D = rng.randn(5,5)
print(D)
```
